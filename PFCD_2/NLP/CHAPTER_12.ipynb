```python
from textblob import TextBlob

text = 'I am sad. I am happy.'
text1 = 'I am happy'
text2 = 'I am neither happy nor sad'

blob = TextBlob(text)
blob1 = TextBlob(text1)
blob2 = TextBlob(text2)

print(blob.sentiment)
print(blob1.sentiment)
print(blob2.sentiment)
```
