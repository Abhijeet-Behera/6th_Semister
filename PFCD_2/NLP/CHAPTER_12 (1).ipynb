```python
from textblob import TextBlob

# Sample texts

text1 = "Love never came to me with a silver spoon, so I licked it off."
text2 = "I was sad. I am happy and elated."
text3 = "I am very happy because I was enjoying my time at home during lockdown."
text4 = """There are vast organizations in Europe and Asia who are rivals. 
The stories of Europe's explorations have reached far and wide. 
Other countries have joined forces with Asia's superior technology 
to level the playing field and change the story."""
text5 = "This product is awful. It smells disgusting. Very disappointing."
# Creating TextBlob objects
blob1 = TextBlob(text1)
blob2 = TextBlob(text2)
blob3 = TextBlob(text3)
blob4 = TextBlob(text4)
blob5 = TextBlob(text5)
# Check sentiment object type
print(type(blob1.sentiment))
print()
# Print sentiments
print("Text1 Sentiment:", blob1.sentiment)
print("Text2 Sentiment:", blob2.sentiment)
print("Text3 Sentiment:", blob3.sentiment)
print("Text4 Sentiment:", blob4.sentiment)
print("Text5 Sentiment:", blob5.sentiment)
# Optional: Print polarity and subjectivity separately
print("\nDetailed Breakdown:")
print("Text1 -> Polarity:", blob1.sentiment.polarity, 
      "Subjectivity:", blob1.sentiment.subjectivity)
print("Text2 -> Polarity:", blob2.sentiment.polarity, 
      "Subjectivity:", blob2.sentiment.subjectivity)
print("Text3 -> Polarity:", blob3.sentiment.polarity, 
      "Subjectivity:", blob3.sentiment.subjectivity)
print("Text4 -> Polarity:", blob4.sentiment.polarity, 
      "Subjectivity:", blob4.sentiment.subjectivity)
print("Text5 -> Polarity:", blob5.sentiment.polarity, 
      "Subjectivity:", blob5.sentiment.subjectivity)
```
